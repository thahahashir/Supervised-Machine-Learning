```python
#PLotting comparision plots between targetvariable and continuous predictors
def PlotBarCharts(inpData, colsToPlot,TargetVariable):
    %matplotlib inline
    
    import matplotlib.pyplot as plt
    
    # Generating multiple subplots
    fig, subPlot=plt.subplots(nrows=1, ncols=len(colsToPlot), figsize=(18,5))
    fig.suptitle('Bar charts of: '+ str(colsToPlot))
    GroupedDataSpecies=creditRiskData.groupby(TargetVariable)
    
    for colName, plotNumber in zip(colsToPlot, range(len(colsToPlot))):
        GroupedDataSpecies.mean()[colName].plot.bar(rot=0,ax=subPlot[plotNumber])
        #inpData.groupby(colName).mean().plot(kind='bar',ax=subPlot[plotNumber])

# Calling the function
inpData=creditRiskData
colsToPlot=QuantitativeVariable
PlotBarCharts(inpData, colsToPlot,TargetVariable)
```

```python
#PLotting comparision plots between targetvariable and predictors
def PlotBoxCharts(inpData, colsToPlot,TargetVariable):
    %matplotlib inline
    
    import matplotlib.pyplot as plt
    # Generating multiple subplots
    fig, subPlot=plt.subplots(nrows=1, ncols=len(colsToPlot), figsize=(18,5))
    fig.suptitle('Box charts of: '+ str(colsToPlot))
    
    for colName, plotNumber in zip(colsToPlot, range(len(colsToPlot))):
        inpData.boxplot(column=colName, by=TargetVariable, figsize=(10,4), vert=True,ax=subPlot[plotNumber])
        
#calling function
inpData=creditRiskData
colsToPlot=QuantitativeVariable
PlotBoxCharts(inpData, colsToPlot, TargetVariable)
###################################################
#Very few outliers so no need to do outliers treatment
```

```python
#Statistical Feature to observe relationship between categorical vs COntinuous data : ANNOVA
# f_oneway() function takes the group data as input and returns F-statistic and P-value
from scipy.stats import f_oneway

# Assumption(H0) is that colName and prices are NOT correlated

# We accept the Assumption(H0) only when P-Value > 0.05
# Here the P-Value is > 0.05 by a slight margin!! which means we will Accept the Assumption(H0)

def FunctionANNOVA(inpData, colsToPlot,TargetVariable):
    
    for colName in (colsToPlot):
        CategoryGroupLists=inpData.groupby(TargetVariable)[colName].apply(list)
        AnnovaResults= f_oneway(*CategoryGroupLists)
        # Suppress the scientific notation in values
        np.set_printoptions(suppress=True)
        print('P-Value for Anova of '+ str(colName) ,'is: ', AnnovaResults[1])

#Calling Function
inpData=creditRiskData
colsToPlot=QuantitativeVariable
FunctionANNOVA(inpData, colsToPlot,TargetVariable)
```

```python
# Cross tablulation plot between two categorical variables
#PLotting comparision plots between targetvariable and predictors categorical categorical

def PlotGrupedBarCharts(inpData, colsToPlot,TargetVariable):
    %matplotlib inline
    
    import matplotlib.pyplot as plt
    # Generating multiple subplots
    fig, subPlot=plt.subplots(nrows=1, ncols=len(colsToPlot), figsize=(18,5))
    fig.suptitle('Bar charts of: '+ str(colsToPlot))
    
    for colName, plotNumber in zip(colsToPlot, range(len(colsToPlot))):
        CrossTabResult=pd.crosstab(index=inpData[colName], columns=inpData[TargetVariable])
        CrossTabResult.plot.bar(figsize=(30,5), color=['red','blue'],ax=subPlot[plotNumber] )
        
#calling function
inpData=creditRiskData
colsToPlot=CategoricalVariables
PlotGrupedBarCharts(inpData, colsToPlot, TargetVariable)

```

```python

# P-Value is the Probability of H0 being True
# If P-Value>0.05 then only we Accept the assumption(H0)


def ChiSquareTest(inpData, colsToPlot,TargetVariable):
    from scipy.stats import chi2_contingency
    
    for colName in (colsToPlot):
        CrossTabResult=pd.crosstab(index=inpData[colName], columns=inpData[TargetVariable])
        # Performing Chi-sq test
        ChiSqResult = chi2_contingency(CrossTabResult)
        print('The P-Value of the ChiSq Test for '+ str(colName) ,'is: ', ChiSqResult[1])
        
#calling function
inpData=creditRiskData
colsToPlot=CategoricalVariables
ChiSquareTest(inpData, colsToPlot, TargetVariable)
```

### Supervised Machine Learning

```python
#Splitting Data into training and testing
Predictors=creditRiskDataNumeric.columns[creditRiskDataNumeric.columns!=TargetVariable]

X=creditRiskDataNumeric[Predictors].values
y=creditRiskDataNumeric[TargetVariable].values

# Split the data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Quick sanity check with the shapes of Training and testing datasets
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## Logistic Regression

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
#choose parameter Penalty='l1' or C=1
# choose different values for solver 'newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'
clf = LogisticRegression(C=1,penalty='l2', solver='newton-cg')

# Printing all the parameters of logistic regression
print(clf)

# Creating the model on Training Data
LOG=clf.fit(X_train,y_train)
prediction=LOG.predict(X_test)

# Measuring accuracy on Testing Data
from sklearn import metrics
print(metrics.classification_report(y_test, prediction))
print(metrics.confusion_matrix(y_test, prediction))

# Printing the Overall Accuracy of the model
F1_Score=metrics.classification_report(y_test, prediction).split()[-2]
print('Accuracy of the model:', F1_Score)

# Printing some sample values of prediction
TestingDataResults=pd.DataFrame(data=X_test, columns=Predictors)
TestingDataResults[TargetVariable]=y_test
TestingDataResults['Predicted'+ TargetVariable]=prediction
TestingDataResults.head()
```

#### Hyper Parameter Tuning

```python
Hyper=pd.DataFrame()
```

```python
def SuperFunction (inpDataNumeric, TargetVariable, Test_Size_Range, Max_Depth_Range, N_Estimator_Range):
    #Empty DataFrame
    HyperParameterTuning=pd.DataFrame()
    for Test_Size in Test_Size_Range:
        #samppling of data
        Predictors=inpDataNumeric.columns[inpDataNumeric.columns !=TargetVariable]
        X=inpDataNumeric[Predictors].values
        y=inpDataNumeric[TargetVariable].values
        #Splitting the data in training and testing
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=Test_Size, random_state=45)
        
        for Max_Depth in Max_Depth_Range :
            for N_Estimator in N_Estimator_Range:
                #print(Test_Size,Max_Depth,N_Estimator)
                
                ####################### Random Forest Regression Model ###############################
                from sklearn.ensemble import RandomForestClassifier
                clf = RandomForestClassifier(max_depth=Max_Depth, n_estimators=N_Estimator,criterion='entropy')

                # Creating the model on Training Data
                RF=clf.fit(X_train,y_train)
                Prediction= RF.predict(X_test)
                
                #Efficiency Metrics
                from sklearn import metrics
                F1_Score = metrics.classification_report(y_test, Prediction).split()[-2]
                HyperParameterTuning=HyperParameterTuning.append(pd.DataFrame(data=[['Random_Forest', Test_Size, Max_Depth,N_Estimator,F1_Score]],
                                                                                      columns=["Model","Test_Size","Max_Depth","N_Estimator","F1_Score"]))
                ########################### Adaboost Model ################################
                from sklearn.ensemble import AdaBoostClassifier
                from sklearn.tree import DecisionTreeClassifier

                # Choosing Decision Tree with 1 level as the weak learner
                DTC=DecisionTreeClassifier(max_depth=Max_Depth)
                clf = AdaBoostClassifier(n_estimators=N_Estimator, base_estimator=DTC ,learning_rate=0.06)
                
                # Creating the model on Training Data
                AB=clf.fit(X_train,y_train)
                Prediction= AB.predict(X_test)
                
                #Efficiency Metrics
                from sklearn import metrics
                F1_Score=metrics.classification_report(y_test, Prediction).split()[-2]
                HyperParameterTuning=HyperParameterTuning.append(pd.DataFrame(data=[['Random_Forest',Test_Size,Max_Depth,N_Estimator,F1_Score]],
                                                                                      columns=["Model","Test_Size","Max_Depth","N_Estimator","F1_Score"]))
                
                ############################ XGBoost Model--> xtreme Gradiant Boost ################################
                from xgboost import XGBClassifier
                clf=XGBClassifier(max_depth=Max_Depth, learning_rate=0.01, n_estimators=N_Estimator, objective='binary:logistic', booster='gbtree')
                
                # Creating the model on Training Data
                XGB=clf.fit(X_train,y_train)
                Prediction= XGB.predict(X_test)      

                #Efficiency Metrics
                from sklearn import metrics
                F1_Score=metrics.classification_report(y_test, Prediction).split()[-2]
                HyperParameterTuning=HyperParameterTuning.append(pd.DataFrame(data=[['Random_Forest',Test_Size,Max_Depth,N_Estimator,F1_Score]],
                                                                                      columns=["Model","Test_Size","Max_Depth","N_Estimator","F1_Score"]))
                
                ############################ KNN--> K Nearest Neighbor ################################
                from sklearn.neighbors import KNeighborsClassifier
                clf = KNeighborsClassifier(n_neighbors=Max_Depth)
                
                # Creating the model on Training Data
                KNN=clf.fit(X_train,y_train)
                Prediction= KNN.predict(X_test)

                #Efficiency Metrics
                from sklearn import metrics
                F1_Score=metrics.classification_report(y_test, Prediction).split()[-2]
                HyperParameterTuning=HyperParameterTuning.append(pd.DataFrame(data=[['Random_Forest',Test_Size,Max_Depth,N_Estimator,F1_Score]],
                                                                                      columns=["Model","Test_Size","Max_Depth","N_Estimator","F1_Score"]))
                
    return HyperParameterTuning.sort_values(by="Model", ascending=False)
   
  
        
```

```python
inpDataNumeric=creditRiskDataNumeric
Test_Size_Range=[0.2,0.25,0.3]
Max_Depth_Range=[2,5,10]
N_Estimator_Range=[100,500]
Hyper=Hyper.append(SuperFunction(inpDataNumeric, TargetVariable, Test_Size_Range, Max_Depth_Range, N_Estimator_Range))
```
